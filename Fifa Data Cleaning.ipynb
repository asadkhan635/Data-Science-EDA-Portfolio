```python
#intitial exploration
#identify some columns that may require cleaning e.g. photoUrl, playerUrl and Team & Contract. These are just a few examples. 
df = pd.read_csv("fifa21_raw_data.csv")

df.head()
```

```python
df['Loan Date End'].unique()
```

```python
#replace blank values with something more intuitive e.g. "No Loan" or "No end date"
df = df.fillna('No Loan')
```

```python
#double check if the code ran successfully
df['Loan Date End'].value_counts()
```

```python
# drop photoUrl and playerUrl columns and then double check if it worked
df = df.drop('playerUrl', axis=1)
df = df.drop('photoUrl', axis=1)

df.head()
```

```python
#Move the ID column up
id_column = df['ID']

df.drop(columns=['ID'], inplace=True)

df.insert(0, 'ID', id_column)
df.head()
```

```python
#you can either use .map with a dictionary or define a function and then use .apply. Both are valid approaches. 
#repear rhe same for D/W score

score_to_points = {'Low': 1, 'Medium': 2, 'High': 3}

df['A/W Score'] = df['A/W'].map(score_to_points)

df.head()
```

```python
#remove the the 'lbs' from the end of the weight column
df['Weight'] = df['Weight'].str[:-3]
df['Weight'] = pd.to_numeric(df['Weight'])


df.head()
```

```python
# we will analyze the age column for outliers

df['Age'].value_counts()
```

```python
# we will use a boolean mask to remove age outliers
df = df[(df['Age'] >= 19) & (df['Age'] < 40)]
df['Age'].value_counts().sort_index()
```

```python
df['Value'] = df['Value'].str[1:]
```

```python
df['Value'].head()
```

```python
#extract the last character to see if they are all in million or if there are others too
last_chars = df['Value'].str[-1]

print(last_chars.value_counts())
```

```python
# for q.9 one option is to define a function and then use .apply
def convert_to_numeric(value):
    if value.endswith('M'):
        return float(value[:-1]) * 1000000
    elif value.endswith('K'):
        return float(value[:-1]) * 1000
    else:
        return float(value)
    
df['Value'] = df['Value'].apply(lambda x: convert_to_numeric(x))
```

```python
df['Value'].head()
```

```python
# another option for q.9 is to use .where from the numpy library
import numpy as np
# Convert 'M' values to numeric
df['Value'] = np.where(df['Value'].str.endswith('M'), 
                       df['Value'].str[:-1].astype(float) * 1000000, 
                       df['Value'])

# Convert 'K' values to numeric
df['Value'] = np.where(df['Value'].str.endswith('K'), 
                       df['Value'].str[:-1].astype(float) * 1000, 
                       df['Value'])

# Convert remaining values to numeric
df['Value'] = df['Value'].astype(float)
```
