```python
#q.6 Which genres/categories have the highest frequency?

genres = dataframe['listed_in'].str.split(',').explode()
genre_counts = genres.str.strip().value_counts()

genre_counts.head()
```

Drama is the most important genre.

```python
#q.7 Find the earliest and latest release year for both TV shows as well as movies.

grouped_dataframe = dataframe.groupby('type') 
grouped_year = grouped_dataframe['release_year']

earliest_years = grouped_year.min()
latest_years = grouped_year.max()

print(latest_years)
print(earliest_years)
```

The earliest TV show was released in 1925 and the latest TV show was releases in 2021. <br> The earliest movie was released in 1942 and the latest movie was releases in 2021. 


```python
#q8. Calculate the percentage of shows or movies added to Netflix before 2010.

dataframe['release_year'] = pd.to_numeric(dataframe['release_year']) #converting datatype
shows_before_2010 = (dataframe['release_year'] < 2010).sum() #aggregating number of shows released before 2010
total_shows = len(dataframe) #getting total number of rows
percentage_before_2010 = (shows_before_2010 / total_shows) * 100
percentage_before_2010
```

15.15% shows were released before 2010.

```python
#q9. Calculate the percentage of TV Shows released in the subcontinent.

#first we will extract the shows and then further shortlist shows that were released in Pak, India and Bangaldesh
#do a quick google research on why we used == instead of = when extracting TV shows (next line)
subcontinent_shows = dataframe[(dataframe['type'] == 'TV Show') & (dataframe['country'].str.contains('India|Pakistan|Bangladesh'))]


subcontinent_shows_count = len(subcontinent_shows)
total_tv_shows = (dataframe['type'] == 'TV Show').sum()
percent_sub_shows = (subcontinent_shows_count / total_tv_shows) * 100

percent_sub_shows
```

16.18% shows originated from Pakistan, India or Bangladesh (subcontinent).
